```python
import os
```

```python
def readrec(file):
    rl=[]
    with open(file) as f:
        for i in f:
            rl.append(i)
    return rl
def readlig(file):
    rl=[]
    with open(file) as f:
        for i in f:
            if i[0]!='H':
                break
            rl.append(i)
    return rl
def hb(recf,ligf,file):
    rec=readrec(recf)
    lig=readlig(ligf)
    with open(file,'w')as f:
        for i in rec:
            print(i.rstrip(),file=f)
        for i in lig:
            print(i.rstrip(),file=f)
    return
```

```python
def list_subfolders(dir_path):
    subfolders = []
    for root, dirs, files in os.walk(dir_path):
        for dir_name in dirs:
            subfolders.append(os.path.join(root, dir_name))
    return subfolders

namel=list_subfolders('./process_pdb/')
print(namel[0])
```

```python
def getnel():
    nel=[]
    with open("noerr.txt") as f:
        for i in f:
            nel.append(i.rstrip())
    return nel
noerrlist=getnel()
```

```python
def getscore(file):
    cnt=0
    score=0.0
    with open(file) as f:
        for i in f:
            #print(cnt,i)
            if cnt==14:
                xi=i.split()
                score=float(xi[1])
                break
            cnt=cnt+1
    return score
```

```python
hbnam=[]
for i in namel:
    rp=i+'/rec.pdb'
    lp=i+'/lig.pdb'
    if rp not in noerrlist:
        continue
    # print(rp,lp)
    # hbp=i+'/hb.pdb'
    # hb(rp,lp,hbp)
    hbnam.append(i)
```

```python
import shutil
import time
def partwork(num):
    time0=time.time()
    st=12*num
    ed=12*(num+1)
    if num==98:
        ed=len(hbnam)
    for i in range(st,ed):
        rel=[]
        outf='./outtm/out'+str(num)
        outs='./outtm/out'+str(num)+'/outsc.txt'
        outt='./outtm/out'+str(num)+'/TM_sup.pdb'
        p0=hbnam[i]
        rep=[]
        rer=[]
        for j in range(len(hbnam)):
            pi=hbnam[j]
            
            r0=p0+'/rec.pdb'
            ri=pi+'/rec.pdb'
            com='./TMalign '+r0+' '+ri+' -o '+outf+'/TM_sup > '+outs
            os.system(com)
            sc=getscore(outs)
            rep.append(pi)
            rer.append(sc)

            if j%200==0:
                print(num,i,j,p0,pi,sc,time.time()-time0)

        outlist=p0+'/tmresult.txt'
        with open(outlist,'w') as f:
            for j in range(len(rep)):
                print(p0,rep[j],rer[j],file=f)

        print(num,i,time.time()-time0)
```
