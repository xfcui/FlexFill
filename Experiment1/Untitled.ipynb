```python
import os

def get_pdb_files(folder_path):
    pdb_files = []
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            if file.endswith('.pdb'):
                pdb_files.append(os.path.join(root, file))
    return pdb_files

# 示例文件夹路径
folder_path = './p450'

# 获取所有的PDB文件
alf = get_pdb_files(folder_path)

# 输出所有的PDB文件
for pdb_file in alf:
    print(pdb_file)
```

```python
import shutil
for i in range(100):
    p='./alfpdb/alf'+str(i)
    os.makedirs(p, exist_ok=True)
    newp=p+'/apdb.pdb'
    shutil.copy(alf[i],p)
    shutil.copy(alf[i],newp)
```

```python
import time
t0=time.time()
for i in range(len(alf)):
    fpath='./alfpdb/alf'+str(i)
    q_pdb=fpath+'/apdb.pdb'
    com='/home/Grx/foldseek/bin/foldseek easy-search '+q_pdb+' ./1000db '+'aln tmpFolder --alignment-type 1 --format-output '+ '"query,target,alntmscore,qtmscore,ttmscore" >fout.txt'
    #print(com)
    os.system(com)
    
    
    results = []
    with open('aln', 'r') as f:
        for line in f:
            parts = line.strip().split()
            query, target, fident, alnlen, TMscore = parts[0], parts[1], float(parts[2]), float(parts[3]), float(parts[4])
            results.append((query, target, fident, alnlen, TMscore))

    # 按 TMscore 降序排序
    results.sort(key=lambda x: x[4], reverse=True)
    
    outl=fpath+'/foldseekout.txt'
    with open(outl,'w') as f:
        for j in results:
            print(j,file=f)
    
    print(i,time.time()-t0)
```

```python
import os

def find_folders_with_cif_files(folder_path):
    folders_with_cif = set()  # 使用集合以避免重复
    for root, dirs, files in os.walk(folder_path):
        if any(file.endswith('.cif') for file in files):
            folders_with_cif.add(os.path.basename(root))  # 添加子文件夹名称
    return folders_with_cif

# 示例文件夹路径
folder_path = './p450'

# 查找包含CIF文件的子文件夹
folderc = find_folders_with_cif_files(folder_path)
print(len(folderc))
```

```python
def get_pdb_files(folder_path):
    pdb_files = []
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            if file.endswith('.pdb'):
                pdb_files.append(os.path.join(root, file))
    return pdb_files
```

```python
folderc=list(folderc)
print(folderc[0])
```

```python
import shutil
for i in range(100):
    p='./alfpdb/alf'+str(i)
    os.makedirs(p, exist_ok=True)
    fp='./p450/'+folderc[i]+'/'
    newp=p+'/apdb.pdb'
    pp=get_pdb_files(fp)
    shutil.copy(pp[0],p)
    shutil.copy(pp[0],newp)
```
